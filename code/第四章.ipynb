```python
# 饼图

from pyecharts.charts import Pie
import pyecharts.options as opts
import pandas as pd

vote_result = pd.read_csv('../data/第四章数据/vote_result.csv')
# data = [list(z) for z in zip(vote_result['Areas_of_interest'], vote_result['Votes'])]

c = (
    Pie()
    .add(
        series_name="饼图",
        data_pair=vote_result.values.tolist(),
        # data_pair=data,
        # label_opts=opts.LabelOpts(is_show=True),
    )
    .set_global_opts(legend_opts=opts.LegendOpts(pos_right='right', pos_top='Top', orient="vertical"))
    .set_series_opts(label_opts=opts.LabelOpts(is_show=True, formatter="{b}: {d}%"))
)

# c.render('./html/vote_result.html')
c.render_notebook()
```

```python
import pandas as pd
from pyecharts.charts import Pie
from pyecharts import options as opts
 
data = pd.read_csv("./data/第四章数据/vote_result.csv")
 
pie = (
       Pie()
       .add('interest',data.values.tolist(),radius=[80,150])
       .set_global_opts(title_opts=opts.TitleOpts(title="用户感兴趣的领域",
                                                  subtitle="以下是读者投票结果",
                                                  pos_left="left",
                                                  pos_top='Top',
                                                  pos_bottom="center"),
                        legend_opts=opts.LegendOpts(pos_right="right",
                                                    orient="vertical"),
       )
)
# pie.render("./h tml/vote_result2.html")
pie.render_notebook()
```

```python
from pyecharts.charts import Bar
import pyecharts.options as opts
import pandas as pd

pre_approval_rate = pd.read_csv('./data/第四章数据/presidential_approval_rate.csv')

datax = list(pre_approval_rate['political_issue'])
datay1 = list(pre_approval_rate['support'])
datay2 = list(pre_approval_rate['oppose'])
datay3 = list(pre_approval_rate['no_opinion'])

bar = (
    Bar()
    .add_xaxis(xaxis_data=datax)
    .add_yaxis("支持", y_axis=datay1, stack=True)  # type: ignore
    .add_yaxis("反对", y_axis=datay2, stack=True)  # type: ignore
    .add_yaxis("不发表意见", y_axis=datay3, stack=True)  # type: ignore
)

# bar.render("./html/presidential_approval_rate.html")
bar.render_notebook()
```

```python
from pyecharts.charts import Line
import pyecharts.options as opts
import pandas as pd

year_population_age = pd.read_csv('./data/第四章数据/us_population_by_age.csv')

line = (
    Line()
    .add_xaxis(xaxis_data=year_population_age['year'].tolist())
    .add_yaxis("5岁以下", y_axis=year_population_age['year_under5'].tolist(), color='orange', stack=True, symbol='diamond', symbol_size=10)  # type: ignore
    .add_yaxis("5至19岁", y_axis=year_population_age['year5_19'].tolist(), color='yellow', stack=True, symbol='triangle', symbol_size=10)  # type: ignore
    .add_yaxis("20至44岁", y_axis=year_population_age['year20_44'].tolist(), color='green', stack=True, symbol='roundRect', symbol_size=10)  # type: ignore
    .add_yaxis("45至64岁", y_axis=year_population_age['year45_64'].tolist(), color='blue', stack=True, symbol='rect', symbol_size=10)  # type: ignore
    .add_yaxis("65岁以上", y_axis=year_population_age['year65above'].tolist(), color='red', stack=True, symbol='circle', symbol_size=10)  # type: ignore
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False),
                     areastyle_opts=opts.AreaStyleOpts(opacity=0.2),
                     markpoint_opts=opts.MarkAreaItem(type_='max'))  # type: ignore
    .set_global_opts(legend_opts=opts.LegendOpts(pos_right='right'),
                     title_opts=opts.TitleOpts(title='人口老龄化'))
)

# line.render('./html/us_population_by_age.html')
line.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.globals import ThemeType
import os
import json
import codecs
from pyecharts.charts import Tree

with codecs.open(os.path.join('./data/第四章数据/GDP_data.json'), 'r', encoding='utf8') as f:
    j = json.load(f)
data = [j]
tree = (
    Tree()
        .add("", data)
        .set_global_opts(title_opts=opts.TitleOpts(title="树图"))
)

tree.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.globals import ThemeType
from pyecharts.charts import TreeMap

with open(os.path.join('./data/第四章数据/GDP_data_1.json'), 'r', encoding='utf8') as f:
    data = json.load(f)
treemap = (
    TreeMap()
    .add("演示数据", data)
    .set_series_opts(label_opts=opts.LabelOpts(is_show=True, position='inside'))
    .set_global_opts(title_opts=opts.TitleOpts(title="矩形树图示例"))
)

treemap.render_notebook()
```
