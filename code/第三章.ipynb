```python
from pyecharts.charts import Line
import pyecharts.options as opts

x = ['1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009']
y = [0.32, 0.32, 0.32, 0.32, 0.33, 0.33, 0.34, 0.37, 0.37, 0.37, 0.37, 0.39, 0.41, 0.42, 0.44]

line = (
    Line()
    .add_xaxis(xaxis_data=x)
    .add_yaxis(
        'Price', 
        y_axis=y,
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="单位：$"),
        yaxis_opts = opts.AxisOpts(
            min_=0.3,
            max_=0.45,
        )
    )
)

line.render_notebook()
# line.render("./html/line.html")
```

```python
import csv
from pyecharts.charts import Line
import pyecharts.options as opts

filename = "./world-population.csv"

datax = []
datay = []

# 读取数据
with open(filename, 'r') as f:
    reader = csv.reader(f)
    for datarow in reader:
        if reader.line_num != 1:
            datax.append(str(datarow[0]))
            datay.append(int(datarow[1]))

line = (
    Line()
    .add_xaxis(xaxis_data=datax)
    .add_yaxis(
        '人口', 
        y_axis=datay,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="历年人口总折线图"))
)

line.render_notebook()
# line.render("./html/world-population.html")
```

```python
from pyecharts.charts import Line,Scatter
import pyecharts.options as opts
import pandas as pd
import numpy as np


df = pd.read_csv('./unemployment-rate-1948-2010.csv',encoding='gbk')

x = df['Year'].astype(str).tolist()
y = df['Value'].tolist()

p = np.polyfit(df['Year'].tolist(),y,deg=2)

#拟合曲线的x,y值
x1 = np.arange(1948,2010)
y1 = np.polyval(p,x1)

#绘制散点图
s = (
    Scatter()
    .add_xaxis(x)
    .add_yaxis(
        "rate-scatter",
        y_axis=y,
        label_opts=opts.LabelOpts(is_show=False),
        symbol_size=5,color='green',
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='1948-2010失业率',
            title_textstyle_opts=opts.TextStyleOpts(color='red')
        )
    )
)

l = (
    Line()
    .add_xaxis(xaxis_data=x1.astype('str').tolist())
    .add_yaxis(
        "rate-line",
        y_axis=y1,
        label_opts=opts.LabelOpts(is_show=False),
        linestyle_opts=opts.LineStyleOpts(width=5),
    )
)

# s.overlap(l).render("./html/unemployment-rate-1948-2010.html")
s.overlap(l).render_notebook()
```

```python
from pyecharts.charts import Scatter
import pyecharts.options as opts
import csv

filename = "./flowingdata_subscribers.csv"

datax = []
datay = []

with open(filename) as f:
    reader = csv.reader(f)
    for datarow in reader:
        if reader.line_num != 1:
            datay.append(int(datarow[1]))

for i in range(1, len(datay) + 1):
    datax.append(str(i))

s = (
    Scatter()
    .add_xaxis(xaxis_data=datax)
    .add_yaxis(
        '订阅量',
         y_axis=datay,
         label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="杂志订阅量"),
        # toolbox_opts=opts.ToolboxOpts(is_show=False),    
    )
)

s.render_notebook()
# s.render("./html/flowingdata_subscribers.html")
```

```python
import pandas as pd
from pyecharts.charts  import Bar
import pyecharts.options as opts
import csv
 
 
datax = []
datay = []
 
filename = './hot-dog-contest-winners.csv'

with open(filename) as f:
    reader = csv.reader(f)
    for datarow in reader:
        if reader.line_num != 1:
            datax.append(datarow[0])
            datay.append(datarow[2])

c = (
    Bar()
    .add_xaxis(xaxis_data=datax)
    .add_yaxis(
        'A',
        y_axis=datay,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(
        yaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=True)),
        title_opts=opts.TitleOpts(title="人口总量(亿)"),
    )
 
)
 
c.render_notebook()
# c.render("./html/hot-dog-contest-winners.html")
```

```python
from pyecharts.charts import Bar
import pyecharts.options as opts
import csv

filename = "./hot-dog-places.csv"

datax = []
datay = []

with open(filename) as f:
    reader = csv.reader(f)
    for datarow in reader:
        datax.append(datarow)

x = datax[0]
y1 = datax[1]
y2 = datax[2]
y3 = datax[3]

bar = (
    Bar()
    .add_xaxis(xaxis_data=x)
    .add_yaxis("A", y_axis=y1, stack=True)
    .add_yaxis("B", y_axis=y2, stack=True)
    .add_yaxis("C", y_axis=y3, stack=True)
    .set_global_opts(title_opts=opts.TitleOpts(title="柱形图数据堆叠示例"))
)

bar.render_notebook()
# bar.render("./html/hot-dog-places.html")
```
